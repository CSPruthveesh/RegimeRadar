```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
depth_files = glob('/content/depth20/*.txt')
aggtrade_files = glob('/content/aggtrade/*.txt')
depth20_df = pd.concat([pd.read_csv(f) for f in depth_files], ignore_index=True)
aggtrade_df = pd.concat([pd.read_csv(f) for f in aggtrade_files], ignore_index=True)
```

```python
import pandas as pd
import re

depth20_df['Timestamp'] = depth20_df['Time'].astype(str).str.extract(r'(\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2})')[0]

depth20_df['Timestamp'] = pd.to_datetime(depth20_df['Timestamp'], errors='coerce')


depth20_df = depth20_df.set_index('Timestamp')
```

```python
depth20_df['Spread'] = depth20_df['AskPriceL1'] - depth20_df['BidPriceL1']
depth20_df['Mid-Price'] = (depth20_df['AskPriceL1'] + depth20_df['BidPriceL1']) / 2
depth20_df['Order Book Imbalance'] = (depth20_df['BidQtyL1'] - depth20_df['AskQtyL1']) / (depth20_df['BidQtyL1'] + depth20_df['AskQtyL1'])

bid_qty_cols = [f'BidQtyL{i}' for i in range(1, 21)]
ask_qty_cols = [f'AskQtyL{i}' for i in range(1, 21)]
depth20_df['Full Depth Order Book Imbalance'] = (depth20_df[bid_qty_cols].sum(axis=1) - depth20_df[ask_qty_cols].sum(axis=1)) / (depth20_df[bid_qty_cols].sum(axis=1) + depth20_df[ask_qty_cols].sum(axis=1))

depth20_df.index = pd.to_datetime(depth20_df.index)
log_returns = np.log(depth20_df['Mid-Price']/depth20_df['Mid-Price'].shift(1))
mean_log_returns = log_returns.rolling(10).mean()
depth20_df['Rolling Volatality'] = mean_log_returns.rolling(10).std()
depth20_df['Rolling Volatality']


```

```python
plt.figure(figsize=(12, 6))
plt.plot(depth20_df.index, depth20_df['Spread'], label='Spread', color='purple')
plt.title('Spread')
plt.xlabel('Time')
plt.ylabel('Spread')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(depth20_df.index, depth20_df['Mid-Price'], label='Mid-Price')
plt.title('Mid-Price')
plt.xlabel('Time')
plt.ylabel('Mid-Price')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(depth20_df.index, depth20_df['Order Book Imbalance'], label='Order Book Imbalance')
plt.title('Order Book Imbalance')
plt.xlabel('Time')
plt.ylabel('Order Book Imbalance')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(depth20_df.index, depth20_df['Rolling Volatality'], label='Rolling Volatality')
plt.title('Rolling Volatality')
plt.xlabel('Time')
plt.ylabel('Rolling Volatality')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
